## Using GroupBy

```python
make_group = df.groupby('Make')

# average price by Make
make_group["Price"].mean()
```

```python
# average mileage by Make
make_group["Mileage"].mean()
```

```python
# Make average price per model using matplotlib 
make_price = make_group["Price"].mean().sort_values(ascending=True)
make_price.plot(kind='barh', ylabel = "Car Make", xlabel = "Average Price", figsize=(10, 58/4), color="Orange");
```

```python
# Make average mileage per model using matplotlib 
make_mileage = make_group["Mileage"].mean().sort_values(ascending=True)
make_mileage.plot(kind='barh', ylabel = "Car Make", xlabel = "Average Mileage", figsize=(10, 58/4), color="Pink");
```

```python
# Make average price per model using seaborn 
make_price = pd.DataFrame(make_group["Price"].mean().sort_values(ascending=False))
f, ax = plt.subplots(figsize=(10, 58/4))
sns.barplot(x="Price", y="Make", data=make_price, color="Orange");
```

```python
# Make average mileage per model using seaborn 
make_mileage = pd.DataFrame(make_group["Mileage"].mean().sort_values(ascending=False))
f, ax = plt.subplots(figsize=(10, 58/4))
sns.barplot(x="Mileage", y="Make", data=make_mileage, color="Pink");
```

## Skatterplot to Compare Specific Price between 2 Models 

```python
honda = df[df['Make']=='Honda']
civic = honda[honda['Model'].str.contains('Civic', na = False)]
```

```python
ford = df[df['Make']=='Ford']
f150 = ford[ford['Model'].str.contains('F-150', na = False)]
```

```python
# with matplotlib
plt.plot(civic["Year"], civic['Price'], 'b.', label='Honda Civic')
plt.plot(f150["Year"],f150["Price"], 'g.', label='Ford F-150')
plt.legend(loc='upper left')
plt.xlabel('Year') 
plt.ylabel('Price') 
plt.title('Honda Civic and Ford F-150 Price by year')
plt.show()
```

```python
# with seaborn
sns.scatterplot(data=civic, x="Year", y="Price", color='blue')
sns.scatterplot(data=f150, x="Year", y="Price", color='green')
plt.legend(labels = ['Honda Civic', 'Ford F-150']);
```

## Splitting the Dataset
Here we split the dataset between North and South States to compare how geographic regions may cause a difference in things such as vehicle price. 

```python
north_states = [' WI', ' NY', ' IL', ' MO',' MT', ' IA', ' ND', ' SD', ' NE', ' MI', ' MN']
north = df[df['State'].isin(north_states)]

south_states = [' TX', ' FL', ' LA', ' NM', ' AZ', ' MS', ' AL', ' GA', ' AR', 'SC', ' NV']
south = df[df['State'].isin(south_states)]

# adding new column to data frame to pick out north or south sample
df['N or S'] = None # ensures object dtype 
df.loc[df['State'].isin(north_states), 'N or S'] = 'N' 
df.loc[df['State'].isin(south_states), 'N or S'] = 'S'

#making sure non selected states are marked null
df.replace('nan', np.nan, inplace=True)

df.head()
```

```python
# getting older models specifically
new_df = df[df["Year"]<2011]
models = ["Civic", 'Sierra', 'Navigator', 'GX', 'Silverado', 'Wrangler']

sample_models = new_df[new_df["Model"].isin(models)]
```

```python
n_sample = sample_models[sample_models["N or S"] == "N"]
s_sample = sample_models[sample_models["N or S"] == "S"]
```

```python
sns.scatterplot(data=n_sample, x="Year", y="Price", color='blue')
sns.scatterplot(data=s_sample, x="Year", y="Price", color='yellow')
plt.legend(labels = ['North', 'South']);
```

The scatter plot shows that for each year neither region has significantly different prices per car. There are some outliers from both sample sets that contain higher prices.

```python
plt.hist(n_sample["Price"], bins=30, alpha=0.5, label='North', color='blue')
plt.hist(s_sample["Price"], bins=30, alpha=0.5, label='South', color='yellow')
plt.xlabel('Price') 
plt.ylabel('No. of Cars') 
plt.title('Car Price Histogram')
plt.legend()
plt.show();
```

While this graph shows that there are more vehicles from the South, we can't really tell whether or not either region's car price is significantly different from the others. 

```python
# Determining Price Per Model in each region to better compare
price_per_model_N = n_sample.groupby('Model')['Price'].mean().reset_index()
price_per_model_S = s_sample.groupby('Model')['Price'].mean().reset_index()

# Display the result
print(price_per_model_N)
print(price_per_model_S)
```

```python
labels = ["Civic", 'Sierra', 'Navigator', 'GX', 'Silverado', 'Wrangler']
sizeN = [7389.707317, 17555.768116, 12383.333333, 15925.000000, 16006.558282, 20717.235294]
sizeS = [7262.728972, 16259.851064, 10499.000000, 17689.555556, 15206.477366, 22382.368421]
colors = ['tab:blue', 'tab:orange', 'tab:green', 'tab:red', 'tab:purple', 'tab:brown', 'tab:pink']

plt.subplot(1,2,1)
plt.pie(sizeN, colors=colors, labels=labels)
plt.title("North Average Price Per Model")

plt.subplot(1,2,2)
plt.pie(sizeS, colors=colors, labels=labels)
plt.title("South Average Price Per Model")

plt.tight_layout()
plt.show()
```

```python
plt.subplot(1,2,1)
plt.barh(price_per_model_N['Model'], price_per_model_N['Price'], color='Blue')
plt.xlabel('Model')
plt.ylabel('Average Price')
plt.title('Average Price per Model North')

plt.subplot(1,2,2)
plt.barh(price_per_model_S['Model'], price_per_model_S['Price'], color='Yellow')
plt.xlabel('Model')
plt.ylabel('Average Price')
plt.title('Average Price per Model South')

plt.tight_layout()
plt.show()
```

Both graphs visualizing the average price per model in the North or South appear to be almost identicle. This suggests that there is not much of a difference between the price for each region

### Conclusion

Based on the several graphs produced above, there appears to be little to no reason to assume that cars in the North depreciate faster than the South. The models appeared to be very similar, especially when comparing average price per model for each region.

## Advanced Seaborn Graphing
The section below is used to show off more advanced abilities of the Seaborn package.

```python
# seeing make counts to only graph some of the largest amounts
df.groupby('Make').count()
```

```python
makes = ['BMW', 'Chevrolet', 'Dodge', 'Ford', 'Honda', 'Hyundai', 'Jeep', 'Kia', 'Nissan', 'Toyota']
make_df = df[df["Make"].isin(makes)]

sns.set_theme(style="darkgrid")

sns.lineplot(x="Year", y="Price",
             hue="Make",
             data=make_df)
#move legend outside of plot
plt.legend(bbox_to_anchor=(1.02, 1), loc='upper left', borderaxespad=0);
```

```python
sns.boxenplot(
    make_df, x="Make", y="Price",
    color="b", width_method="linear",
)
plt.tight_layout()
```

### Explaination of Plots

The two seaborn plots above picked the 10 car makes with the most observations (rows) in the data set. The first plot then graphs the year by price fo each car make on the same graph so they can easily be compared against each other over time. The second plot shows a boxenplot or letter-value plot for each car make's price. A letter-value plot is very similar to a boxplot, except it is built for larger data sets such as this one as it boxes for more percentiles. This better shows the distribution of each data set and the several price outliers (the dots) for each model.
